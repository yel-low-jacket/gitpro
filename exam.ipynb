```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Евгеника Тиктоник\Downloads\exam.csv')
df
```

```python
sorted_data1 = df[(df['TG'] == 0) & (df['MIP'] >= 63.90625) & (df['MIP'] <= 175.328125)]

sorted_data2 = df[(df['TG'] == 1) & (df['MIP'] >= 8.15625) & (df['MIP'] <= 119.671875)]

sorted_data = pd.concat([sorted_data1, sorted_data2], ignore_index=True, sort=False)
sorted_data
```

```python
sorted_data['MC'].mean()
```

```python
ascend = sorted_data.sort_values(by=['SIP'])
```

```python
from sklearn.model_selection import train_test_split
def build_model(df):
    X, y = df.drop('TG', axis = 1), df['TG']
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 17, test_size = 0.2, stratify = y)
    return X_train.SC.mean()
mean = build_model(ascend)
mean
```

```python
from sklearn.preprocessing import MinMaxScaler
def build_model(df):
    X, y = df.drop('TG', axis = 1), df['TG']
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 17, test_size = 0.2, stratify = y)
    scaler = MinMaxScaler().fit_transform(X_train)
    return scaler.EKC.mean()
mean = build_model(ascend)
mean
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
def build_model(df):
    X, y = df.drop('TG', axis = 1), df['TG']
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 17, test_size = 0.2, stratify = y)
    scaler = MinMaxScaler().fit(X_train)
    X_train_pr = scaler.transform(X_train)
    lr = LogisticRegression(random_state = 17).fit(X_train_pr, y_train)
    y_pred = lr.predict(X_test)
    f_1 = f1_score(y_test, y_pred)
    return f_1
score = build_model(ascend)
score
```

```python
from sklearn.neighbors import KNeighborsClassifier
def build_model(df):
    X, y = df.drop('TG', axis = 1), df['TG']
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 17, test_size = 0.2, stratify = y)
    scaler = MinMaxScaler().fit_transform(X_train)
    knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    f_1 = f1_score(y_test, y_pred)
    return f_1
score = build_model(ascend)
score
```
